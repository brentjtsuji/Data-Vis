```python
import pandas as pd
import numpy as np
import seaborn as sns
import keras
```

```python
from keras.datasets import boston_housing
```

```python
(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(train_data)
```

```python
df = df.rename(columns = {0:'cr',1:'sf',2:'nr',3:'river',4:'nitric',5:'avg',6:'own',7:'dist',8:'hwy',9:'tax',10:'teach',11:'blk',12:'lwr'})
```

```python
df.head()
```

```python
sns.lmplot(x='nr', y='sf', data =df)
```

```python
sns.lmplot(x='nr', y='sf', data =df,
          fit_reg=False,
          hue = 'river')
```

```python
sns.boxplot(data = df)
```

```python
sns.set_style('whitegrid')
sns.violinplot(data = df)
```

```python
sns.swarmplot(x='nr', y='sf',data=df)
```

```python
corr = df.corr()
sns.heatmap(corr)
```
